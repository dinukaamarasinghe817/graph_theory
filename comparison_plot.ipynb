```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('output.csv')

# define the x-axis values (edges)
x = df['edges']

algorithms = ['kruskals', 'lazyprims', 'eagerprims']
colors = ['b', 'r']

# iterate through each algorithm and create a separate plot
for algorithm in algorithms:
    fig, ax = plt.subplots(figsize=(10, 6))
    
    ax.set_xscale('log')
    ax.set_yscale('log')
    
    for color, representation in zip(colors, ['adj_list', 'adj_mat']):
        col = f'{algorithm}_{representation}'
        label = f'{algorithm} ({representation})'
        y = df[col]
        
        ax.plot(x, y, label=label, color=color, marker='o', markersize=3)
    
    ax.set_xlabel('Edges (log scale)')
    ax.set_ylabel('Execution Time (log scale)')
    ax.set_title(f'Execution Times for {algorithm.capitalize()} Algorithm (Adj List vs. Adj Matrix)')
    
    ax.legend(fontsize='small')
    
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.show()

# all together
columns = ['kruskals_adj_list', 'kruskals_adj_mat', 'lazyprims_adj_list', 'lazyprims_adj_mat', 'eagerprims_adj_list', 'eagerprims_adj_mat']
colors = ['b', 'g', 'r', 'c', 'm', 'y']

fig, ax = plt.subplots(figsize=(10, 6))

ax.set_xscale('log')
ax.set_yscale('log')

for col, color in zip(columns, colors):
    y = df[col]
    ax.plot(x, y, label=col, color=color, marker='o', markersize=3)

ax.set_xlabel('Edges (log scale)')
ax.set_ylabel('Execution Time (log scale)')
ax.set_title('Execution Times for Different Algorithms')

ax.legend(fontsize='small')

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()

# Print execution times
print(df)
```
